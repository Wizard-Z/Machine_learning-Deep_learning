## Census file of Baby names year 2017

```python
import pandas as pd
```

```python
df = pd.read_csv('yob2017.txt',header = None)
df.head()
```

## Girls

```python
f_first = df[df[1] == 'F'][0].apply(lambda x:x[0])
f_second = df[df[1] == 'F'][0].apply(lambda x:x[1])
f_last = df[df[1] == 'F'][0].apply(lambda x:x[-1])
f_second_last = df[df[1] == 'F'][0].apply(lambda x:x[-2])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualizations
### Female names

```python
fig, axs = plt.subplots(nrows=2,figsize = (15,12))
sns.countplot(f_first,ax = axs[0])
axs[0].title.set_text('First Character in Girl\'s Names')
sns.countplot(f_second,ax=axs[1])
axs[1].title.set_text('Second Character in Girls\'s Names')
axs[1].set_xlabel('Alphabets')
#plt.savefig('Girls First Character.jpg',dpi = 300)
```

```python
fig, axs = plt.subplots(nrows=2,figsize = (15,12))
sns.countplot(f_last,ax = axs[0])
axs[0].title.set_text('Last Character in Girl\'s Names')
sns.countplot(f_second_last,ax=axs[1])
axs[1].title.set_text('Second Last Character in Girls\'s Names')
axs[1].set_xlabel('Alphabets')
#plt.savefig('Girls last Character.jpg',dpi = 300)
```

### Boys

```python
m_first = df[df[1] == 'M'][0].apply(lambda x:x[0])
m_second = df[df[1] == 'M'][0].apply(lambda x:x[1])
m_last = df[df[1] == 'M'][0].apply(lambda x:x[-1])
m_second_last = df[df[1] == 'M'][0].apply(lambda x:x[-2])
```

```python
fig, axs = plt.subplots(nrows=2,figsize = (15,12))
sns.countplot(m_first,ax = axs[0])
axs[0].title.set_text('First Character in Boys\'s Names')
sns.countplot(m_second,ax=axs[1])
axs[1].title.set_text('Second Character in Boys\'s Names')
axs[1].set_xlabel('Alphabets')
#plt.savefig('Boys First Character.jpg',dpi = 300)
```

```python
fig, axs = plt.subplots(nrows=2,figsize = (15,12))
sns.countplot(m_last,ax = axs[0])
axs[0].title.set_text('Last Character in Boys\'s Names')
sns.countplot(m_second_last,ax=axs[1])
axs[1].title.set_text('Second Character in Boys\'s Names')
axs[1].set_xlabel('Alphabets')
#plt.savefig('Boys Last Character.jpg',dpi = 300)
```

```python
fig, axs = plt.subplots(nrows=4,ncols=2,figsize = (17,30))
sns.countplot(f_first,ax = axs[0][0])
sns.countplot(f_second,ax = axs[0][1])
sns.countplot(m_first,ax = axs[1][0])
sns.countplot(m_second,ax = axs[1][1])
sns.countplot(f_last,ax = axs[2][0])
sns.countplot(f_second_last,ax = axs[2][1])
sns.countplot(m_last,ax = axs[3][0])
sns.countplot(m_second_last,ax = axs[3][1])

axs[0][0].title.set_text('First Character in Girls\'s Names')
axs[0][1].title.set_text('Second Character in Girls\'s Names')
axs[1][0].title.set_text('First Character in Boys\'s Names')
axs[1][1].title.set_text('Second Character in Boys\'s Names')
axs[2][0].title.set_text('Last Character in Girls\'s Names')
axs[2][1].title.set_text('Second Last Character in Girls\'s Names')
axs[3][0].title.set_text('Last Character in Boys\'s Names')
axs[3][1].title.set_text('Second Last Character in Boys\'s Names')

#axs.set_xlabel('Alphabets')
```
