```python
test1 = Image.open('t1.jpg')
test2 = Image.open('t2.jpg')
test3 = Image.open('t3.jpg')
test4 = Image.open('t4.jpg')
```

```python
r1 = pytesseract.image_to_string(test1)
r2 = pytesseract.image_to_string(test2)
r3 = pytesseract.image_to_string(test3)
r4 = pytesseract.image_to_string(test4)
```

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (20,20))
axs[0][0].imshow(test1)
axs[0][0].axis('off')
axs[0][0].set_title(r1)

axs[0][1].imshow(test2)
axs[0][1].axis('off')
axs[0][1].set_title(r2)

axs[1][0].imshow(test3)
axs[1][0].axis('off')
axs[1][0].set_title(r3)

axs[1][1].imshow(test4)
axs[1][1].axis('off')
axs[1][1].set_title(r4)
;
```
